```python
import nltk
import glob, os
import string
import numpy as np
import copy
import pandas as pd
import pickle
import re
import math
```

```python
import glob, os
data = []
listFile = [];
for file in glob.glob("../*.csv"):
    try:
        df = pd.read_csv(file)
        data.append([x.lower() for x in list (df['Operation'].values)])
        listFile.append(file)
    except:
        print(file)
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
```

```python
tagged_data = [TaggedDocument(words=_d, tags=[str(i)]) for i, _d in enumerate(data)]
```

```python
tagged_data[0]
```

```python
max_epochs = 100
vec_size = 100
alpha = 0.025

model = Doc2Vec(size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =0)
  
model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))bl
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("d2v.model")
print("Model Saved")
```

```python
from gensim.models.doc2vec import Doc2Vec

model= Doc2Vec.load("d2v.model")
```

```python
len(model.docvecs)
```

```python
from scipy import spatial

n = len(model.docvecs)
simarray = []
for i in range(0,n):
    narray = []
    for j in range(0,n):
        if (i!=j):
            result = 1 - spatial.distance.cosine(model.docvecs[i] , model.docvecs[j])
            narray.append({"id":j,"key":listFile[j],"value":result})
    simarray.append({"id":i,"key":listFile[i],"value":narray})
```

```python
import json
with open('sim.json', 'w') as outfile:
    json.dump(simarray, outfile)
```
